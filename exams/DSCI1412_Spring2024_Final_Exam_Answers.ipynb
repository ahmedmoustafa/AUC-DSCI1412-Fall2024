## Question 1

### Problem Statement

A box contains 2 gold coins, 3 silver coins, and 4 bronze coins. A coin is drawn randomly and not replaced. Then a second coin is drawn. Calculate the probability of drawing 2 silver coins in a row.

### Solution

* **Step 1: Calculate the probability of drawing one silver coin.**

There are 3 silver coins and a total of 9 coins. Therefore, the probability of drawing one silver coin is  $\frac{3}{9} = \frac{1}{3}$

* **Step 2: Calculate the probability of drawing another silver coin after the first one is not replaced.**

After taking out one silver coin, there are only 2 silver coins left and a total of 8 coins. Therefore, the probability of drawing another silver coin is  $\frac{2}{8} = \frac{1}{4}$

* **Step 3: Calculate the probability of drawing two silver coins in a row.**

To get the probability of both events happening, we multiply the probabilities of each event. Therefore, the probability of drawing 2 silver coins in a row is  $\frac{1}{3} \times \frac{1}{4} = \frac{1}{12}$

## Question 2

### Problem Statement

A fair coin is flipped 10 times. Calculate the probability of getting exactly 6 heads.

### Solution

* **Step 1: Calculate the number of possible outcomes.**

Each coin flip has 2 possible outcomes (heads or tails). Therefore, flipping a coin 10 times has $2^{10} = 1024$ possible outcomes.

* **Step 2: Calculate the number of favorable outcomes.**

The number of ways to get exactly 6 heads in 10 flips is given by the binomial coefficient:  $^{10}C_6 = \frac{10!}{6!(10-6)!} = 210$

* **Step 3: Calculate the probability.**

The probability of getting exactly 6 heads is the number of favorable outcomes divided by the total number of possible outcomes:  $\frac{210}{1024} = 0.205$

## Question 3

### Problem Statement

The heights of students in a university are normally distributed with a mean of 170 cm and a standard deviation of 10 cm. Find the probability that a randomly selected student is taller than 180 cm.

### Solution

* **Step 1: Standardize the value.**

We need to standardize the value of 180 cm using the z-score formula:
$$z = \frac{x - \mu}{\sigma}$$
where:
- $x$ is the value we want to standardize (180 cm)
- $\mu$ is the mean (170 cm)
- $\sigma$ is the standard deviation (10 cm)

Therefore,  $z = \frac{180 - 170}{10} = 1$

* **Step 2: Find the probability.**

We can use a z-table to find the probability that a standard normal random variable is greater than 1. This probability is 0.1587.

Therefore, the probability that a randomly selected student is taller than 180 cm is 0.1587.

## Question 5

### Problem Statement

A medical test for a particular disease has a sensitivity (true positive rate) of 95% and a specificity (true negative rate) of 90%. In a certain population, it is estimated that 2% of individuals have this disease.

a. Calculate the probability that a randomly selected individual who tests positive actually has the disease.

b. Calculate the probability that a randomly selected individual who tests negative does not have the disease.

### Solution

We can use Bayes' theorem to solve this problem. Let's define the following events:

* $D$: the event that an individual has the disease.
* $T$: the event that an individual tests positive.

We are given the following probabilities:

* $P(D) = 0.02$ (the prevalence of the disease)
* $P(T|D) = 0.95$ (the sensitivity of the test)
* $P(T^c|D^c) = 0.90$ (the specificity of the test)

a. **Calculate the probability that a randomly selected individual who tests positive actually has the disease.**

We want to find $P(D|T)$. Using Bayes' theorem, we have:

$P(D|T) = \frac{P(T|D)P(D)}{P(T)}$

To find $P(T)$, we can use the law of total probability:

$P(T) = P(T|D)P(D) + P(T|D^c)P(D^c)$

We can find $P(T|D^c)$ as follows:

$P(T|D^c) = 1 - P(T^c|D^c) = 1 - 0.90 = 0.10$

Similarly, we can find $P(D^c)$ as follows:

$P(D^c) = 1 - P(D) = 1 - 0.02 = 0.98$

Substituting these values into the equation for $P(T)$, we get:

$P(T) = (0.95)(0.02) + (0.10)(0.98) = 0.117$

Substituting these values into the equation for $P(D|T)$, we get:

$P(D|T) = \frac{(0.95)(0.02)}{0.117} = 0.162$

Therefore, the probability that a randomly selected individual who tests positive actually has the disease is 0.162.

b. **Calculate the probability that a randomly selected individual who tests negative does not have the disease.**

We want to find $P(D^c|T^c)$. Using Bayes' theorem, we have:

$P(D^c|T^c) = \frac{P(T^c|D^c)P(D^c)}{P(T^c)}$

To find $P(T^c)$, we can use the law of total probability:

$P(T^c) = P(T^c|D)P(D) + P(T^c|D^c)P(D^c)$

We can find $P(T^c|D)$ as follows:

$P(T^c|D) = 1 - P(T|D) = 1 - 0.95 = 0.05$

Substituting these values into the equation for $P(T^c)$, we get:

$P(T^c) = (0.05)(0.02) + (0.90)(0.98) = 0.883$

Substituting these values into the equation for $P(D^c|T^c)$, we get:

$P(D^c|T^c) = \frac{(0.90)(0.98)}{0.883} = 0.999$

Therefore, the probability that a randomly selected individual who tests negative does not have the disease is 0.999.

```
def process_values(numbers):
    results = []
    for i in range(len(numbers)):
        if i % 2 == 0:
            results.append(numbers[i] + 1)
        else:
            results.append(numbers[i] - 1)
    return results
```

|5|1|Odd|5-1|4|

Therefore, the output of the code for the list [2, 5, 7, 12, 15] is [3, 4, 8, 11, 16].

## Question 8

### Problem Statement

Write a Python function `find_divisors(n)` that returns a list of all divisors of a given integer `n`.

### Solution

```
def find_divisors(n):
    """
    Returns a list of all divisors of a given integer n.
    """
    divisors = []
    for i in range(1, n + 1):
        if n % i == 0:
            divisors.append(i)
    return divisors
```

```
find_divisors(10)
```

## Question 9

### Problem Statement

Write a recursive Python function `sum_digits(n)` that returns the sum of the digits of a given integer `n`.

### Solution

```
def sum_digits(n):
    """
    Returns the sum of the digits of a given integer n.
    """
    if n < 10:
        return n
    else:
        return n % 10 + sum_digits(n // 10)
```

```
sum_digits(123)
```
